## Problem: The dataset of student information contains duplicate entries of students


possible reasons: students have different accounts when switching grades/schools, second issue is that they may be listed as inactive in one instance but the inactive indicator is missing in another instance. 

To Resolve this issue itterate throught the dataset and assign inactive indicators for every instance of the students ID's and remove them based on the active/inactive status

##### if there are any remaining duplicates investigate them further

#### Make list of unique Student ID's

```python
unique_ids = []

for i in range(len(df)):
    if df.iloc[i,0] not in unique_ids:
        unique_ids.append(df.iloc[i,0])
    
```

#### Create list of inactive indicators

```python
inactive_indicators = []

for i in range(len(df)):
    
    if df.iloc[i,13] not in inactive_indicators:
        inactive_indicators.append(df.iloc[i,13])
```

```python
inactive_indicators
```

##### Remove NaN from list of inactive idicators

```python
inactive_indicators.remove(np.nan)
```

#### Create list of Inactive ID's

```python
inactive_ids = []

for i in range(len(df)):
    
    if df.iloc[i,13] in inactive_indicators:
        inactive_ids.append(df.iloc[i,0])
```

```python
print(f"There are {len(inactive_ids)} students who are indicated as inactive")
```

#### Create inactive_stu column and populate indicating "Inactive Student"

```python
df['inactive_stu'] = ""
```

```python
for i in range(len(df)):
    
    if df.iloc[i,0] in inactive_ids:
        df.iloc[i,35] = 'inactive_student'
```

### Create id_count column and populate indicating number of instances of the ID in the dataset

```python
df['id_count'] = np.nan
```

```python

for i in range(len(df)):
    
    id_count = len(df[df["Student ID"] == df.iloc[i,0]])                              
    df.iloc[i,36] = id_count
```
