```python
import requests
link = "https://en.wikipedia.org/wiki/List_of_highest-grossing_films"
link_text = requests.get(link).text
print(link_text)

```

```python
from bs4 import BeautifulSoup

converted_text = BeautifulSoup(link_text, 'html.parser')
print(converted_text.prettify())
```

```python
print(converted_text.title.string)
```

```python
converted_text.find_all('a')
```

```python
#Fetch all the table tags
tables = converted_text.find_all('table')
print(tables)
```

```python
#fetch all the table tags with class name="wikitable sortable"
main_table = converted_text.find('table', class_= 'wikitable sortable plainrowheaders')
print(main_table)
```

```python
#In the table that we will fetch find the <a> </a>tags  
table_links = main_table.find_all('a')
print(table_links)
```

```python
#put the title into a list 
movies = []
for links in table_links:
    movies.append(links.get('title'))
print(movies)
```

```python
#Convert the list into a dataframe 
import pandas as pd
df = pd.DataFrame(movies)
print(df.head())
```
