## Basic Level

### Part 1: Read in the file with `csv.reader()` and store it in an object called `file_nested_list`.

Hint: This is a TSV (tab-separated value) file, and `csv.reader()` needs to be told [how to handle it](https://docs.python.org/2/library/csv.html).

```python
import csv
import pandas as pd
import numpy as np

from collections import namedtuple   # Convenient to store the data rows

DATA_FILE = './data/chipotle.tsv'
```

```python
# pandas import 
# df = pd.read_csv(file_path, sep='\t', header=0)


# python csv import
file_nested_list = []

with open(DATA_FILE, 'rU') as f:
    rows = csv.reader(f, delimiter = '\t')
    
    for row in rows:
        file_nested_list.append(row)
        
f.close()
```

### Part 2: Separate `file_nested_list` into the `header` and the `data`.


```python
header = file_nested_list[0]
data = file_nested_list[1:]
```

```python
df = pd.DataFrame(data)
df.columns = header
```

```python
df.head()
# df.dtypes
```

## Intermediate Level

### Part 3: Calculate the average price of an order.

Hint: Examine the data to see if the `quantity` column is relevant to this calculation.

Hint: Think carefully about the simplest way to do this!

```python
# Clean Item Price Series
# becasue "quantity" is an object/string data type - convert the series to integers


# remove dollar sign from item_price and convert to float
df['item_price'] = df.item_price.apply(lambda x: x.replace("$", ""))
df['item_price'] = df.item_price.apply(lambda x: float(x))


# Clean Quantity Series
df['quantity'] = df.quantity.apply(lambda x: int(x))


# Clean Order ID Series
df['order_id'] = df.order_id.apply(lambda x: int(x))

```

```python
# to account for the quantity of a given order - create a new column which takes the product of quantity and price

df['quantity*price'] = df.quantity * df.item_price
```

```python
# grouping by order id and sum by order id
order_sum_stat = df.groupby('order_id').sum().mean()
avg_order_total = order_sum_stat[2]




print(f'Using pandas: \n On average, each order totals: ${round(avg_order_total, 2)}\n')

# Using a for loop itterate over the dataframe 
# to find each sum for each order id, appending each order total to an intermediate list

curr_id = 1
curr_tot = 0
result = []

for i in range(len(df.order_id)):
    
    if curr_id == df.iloc[i,0]:
        curr_tot += df.iloc[i,4] * df.iloc[i,1]
    else:
        result.append(round(curr_tot, 2))
        curr_tot = df.iloc[i,4] * df.iloc[i,1]
        curr_id += 1
             
result.append(round(curr_tot,2))



mean_order_total = 0
tot = 0
count = 0
for i in result:
    tot += i
    count += 1
    mean_order_total = tot/count
    
    

print(f"Using a for loop: \n On average, each order totals: ${round(mean_order_total, 2)} \n")

print(f"Using NumPy: \n On average, each order totals: ${round(np.mean(result), 2)}\n")
        
```

### Part 4: Create a list (or set) named `unique_sodas` containing all of unique sodas and soft drinks that Chipotle sells.

Note: Just look for `'Canned Soda'` and `'Canned Soft Drink'`, and ignore other drinks like `'Izze'`.

```python
# replace coca cola with coke since there are more orders that take the description of coke and because they are the same
df['choice_description'] = df['choice_description'].apply(lambda x: x.replace('Coca Cola', 'Coke'))
```

```python
# define mask for item names that match the criteria for sodas or soft drinks
soda_mask = df['item_name'].str.contains('Canned')
```

```python
# assign unique list of canned soda or soft drinks to a numpy array
soda_array = df[soda_mask]['choice_description'].unique()
```

```python
# define empty list UNIQUE SODAS
unique_sodas = []

for i in soda_array:
    soda = i.replace("[", "").replace(']', "")
    
    if soda not in unique_sodas:
        unique_sodas.append(soda)
        
        
unique_sodas.sort()
print(f"The alphabetically sorted list of unique sodas and softdrinks offered at Chipotle: \n {unique_sodas}")

    
```

## Advanced Level


### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
# because the choice description values are type str, remove the brackets to clean the data in each cell

df['choice_description'] = df.choice_description.apply(lambda x: x.replace("[", "").replace("]", ""))
```

```python
# create mask to filter dataFrame by burrito

all_burrito_mask = df['item_name'].str.contains('Burrito')
```

```python

result = []
for i in df[all_burrito_mask]['choice_description']:
    count = 1
    for j in i:
        if j == ",":
            count += 1
    result.append(count)
    
avg_topping_count = np.mean(result)

print(f"Using for loops, \nAnd counting the number of commas as a proxy for the numebr of toppings, \nthe mean value of topping per order is: {round(avg_topping_count)}")
    
    
    
    
    
    
# the following assigns the list "result" to a copy of the dataframe as a column
burrito_df = df[all_burrito_mask].assign(topping_count = result)

print(f"\n\nUsing pandas Assign function to assign the list created in the loops to a dataFrame, \nThe average number of toppings per burrito is: {round(burrito_df['topping_count'].mean())}")
   
```

### Part 6: Create a dictionary. Let the keys represent chip orders and the values represent the total number of orders.

Expected output: `{'Chips and Roasted Chili-Corn Salsa': 18, ... }`

Note: Please take the `quantity` column into account!

Optional: Learn how to use `.defaultdict()` to simplify your code.

```python
# remove hyphens from strings to get a more accurate count of item_names
df['item_name'] = df['item_name'].apply(lambda x: x.replace("-", " "))
```

```python
# create mask of for data containing chips in item names
mask_chips = df['item_name'].str.contains('Chips')

# create series of masked chip dataframe
chip_series = df[mask_chips].item_name
```

```python
# create dicitionary
chips_dict = {}


# itterate over set of unique chip orders
for i in chip_series.unique():
    count = 0
    
# nested itteration of all chip item_named series and check if 
    for j in range(len(chip_series)):
        if i == df[mask_chips].iloc[j, 2]:
            count += df[mask_chips].iloc[j, 1]
            
            
# define chip dictionary key and value
    chips_dict[i] = count
    
print(f" \n\n Using for loops to itterate over the series of chips the following dictionary results:")

chips_dict
```

```python
# using groupyby method, finding the sum of item names and the to_dict method 
print("\n\n Using multiple pandas methods, a more efficient process, creates the following dictionary:")

df[mask_chips].groupby("item_name").sum()['quantity'].to_dict()
```
